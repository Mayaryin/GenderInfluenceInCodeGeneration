# Translation

```python
import sqlite3
import pandas as pd
import os

db_path = "../../data/giicg.db"
if not os.path.exists(db_path):
    raise FileNotFoundError(f"Database file does not exist: {db_path}")

conn = sqlite3.connect(db_path)
prompts = pd.read_sql("SELECT * FROM scratch_prompts", conn)
prompts
```

```python
from helpers.translation import load_language_detection_model, detect_language

tokenizer, lang_model = load_language_detection_model()
prompts['language'] = prompts['message_text'].apply(detect_language, args=(tokenizer, lang_model))
prompts
```

```python
from helpers.translation import load_model, conditional_translation

de_tokenizer, de_model = load_model("Helsinki-NLP/opus-mt-de-en")
it_tokenizer, it_model = load_model("Helsinki-NLP/opus-mt-it-en")
prompts['message_text'] = prompts.apply(
    lambda row: conditional_translation(row, de_tokenizer, de_model, it_tokenizer, it_model, 'message_text'), axis=1
)

prompts
```

```python
prompts.to_sql("translated_scratch_prompts", conn, if_exists="replace", index=False)
conn.close()
```
