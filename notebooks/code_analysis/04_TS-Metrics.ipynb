```python
import sqlite3
import pandas as pd
import os

db_path = "../../giicg.db"
if not os.path.exists(db_path):
    raise FileNotFoundError(f"Database file does not exist: {db_path}")

conn = sqlite3.connect(db_path)
snippets = pd.read_sql("SELECT * FROM translated_scratch_prompts_with_code", conn)
snippets = snippets.dropna(subset=['code'])
snippets = snippets[snippets['code'].str.strip() != '']
snippets = snippets.dropna(subset=['code'])
snippets = snippets[(snippets['gender'] == 'Woman (cisgender)') | (snippets['gender'] == 'Man (cisgender)')]
snippets = snippets[(snippets['programming_language'] != 'python')]
snippets
```

## Clean out html and css

```python
js_only_114 = """function reloadIframe() {
  var iframe = document.getElementById('myIframe');
  var src = iframe.src.split('?')[0]; // Remove existing query parameters, if any
  iframe.src = src + '?reload=' + new Date().getTime(); // Add timestamp to force reload
}
"""

js_only_12 = """function reloadIframe() {
  const iframe = document.getElementById('myIframe');
  // Strip existing query string (optional)
  const url = iframe.src.split('?')[0];
  iframe.src = url + '?t=' + new Date().getTime();
}
setInterval(reloadIframe, 5000); // Reload every 5 seconds
// imagine this is where you detect update completion
onFileUpdateComplete(() => {
  reloadIframe();
});
"""
js_only_80 = """function reloadIframe() {
  const iframe = document.getElementById("myFrame");
  // Option 1: reset src (will reload)
  iframe.src = iframe.src;

  // If caching causes stale content, add a dummy cache-buster
  // iframe.src = iframe.src.split('?')[0] + '?t=' + new Date().getTime();
}
"""
js_only_46 = """function reloadFrame() {
    var frame = document.getElementById('liveFrame');

    /* ------------- option 1: simple reload ------------- */
    // Works when the frame is on the same origin as the parent page.
    // true → bypass cache in most browsers
    frame.contentWindow.location.reload(true);

    /* ------------- option 2: change the src ------------- */
    // Works even for cross-origin frames and always bypasses cache.
    // Keeps the part before the ? and appends a unique time stamp.
    // (You can comment option 1 out and use only this if you prefer.)
    var urlWithoutQuery = frame.src.split('?')[0];
    frame.src = urlWithoutQuery + '?_=' + Date.now();
}
"""
snippets.loc[snippets['code_reply_id'] == 114, 'code'] = js_only_114
snippets.loc[snippets['code_reply_id'] == 12, 'code'] = js_only_12
snippets.loc[snippets['code_reply_id'] == 80, 'code'] = js_only_80
snippets.loc[snippets['code_reply_id'] == 46, 'code'] = js_only_46

snippets.loc[snippets['programming_language'] == 'html_js', 'programming_language'] = "javascript"

snippets = snippets[(snippets['programming_language'] != 'html_css')]
snippets
```

## Export code snippets as files

```python
output_dir = "tsmetrics_snippets"
os.makedirs(output_dir, exist_ok=True)

for i, row in snippets.iterrows():
    code = row["code"]
    lang = row.get("programming_language")
    ext = "ts" if lang == "typescript" else "js"
    with open(f"{output_dir}/snippet_{row['code_reply_id']}.{ext}", "w") as f:
        f.write(code)
```
