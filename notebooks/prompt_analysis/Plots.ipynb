```python
import sqlite3
from helpers.stats import make_annotated_code_blocks_query
import matplotlib.pyplot as plt
import pandas as pd

conn = sqlite3.connect('../../giicg.db')
plt.style.use('ggplot')
```

## Code Blocks per User

```python
query = make_annotated_code_blocks_query()
code_blocks_df = pd.read_sql_query(query, conn)

ax = code_blocks_df["user_id"].value_counts().sort_index().plot(kind="bar", title="Code Blocks per User", figsize=(12, 8))
ax.set_xlabel("User ID")
ax.set_ylabel("Number of Code Blocks")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Messages per User


```python
query = "Select * from messages_annotated"
messages_df = pd.read_sql_query(query, conn)

ax = messages_df["user_id"].value_counts().sort_index().plot(kind="bar", title="Messages per User", figsize=(12, 8))
ax.set_xlabel("User ID")
ax.set_ylabel("Number of Code Blocks")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
