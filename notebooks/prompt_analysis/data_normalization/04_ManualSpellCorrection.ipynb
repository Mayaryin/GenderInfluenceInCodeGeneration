```python
def apply_all_rules(df):
    """Apply all cleaning rules in sequence"""
    print(f"Starting with {len(df)} rows")
    print("-" * 40)

    df = fix_spelling_errors(df)

    print("-" * 40)
    print(f"Final: {len(df)} rows")
    return df

# Apply rules
corrected_prompts = apply_all_rules(prompts.copy())
corrected_prompts
```
